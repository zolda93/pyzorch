```python
import zorch
from zorch import nn
```

```python
m = nn.AvgPool2d((2, 3), stride=(1, 2),padding=(1,1))
s = nn.Sigmoid()
```

```python
input = zorch.Tensor([[[[0.8859, 0.7493, 0.9795, 0.5785],
          [0.3848, 0.0331, 0.4285, 0.2467],
          [0.5630, 0.2204, 0.8608, 0.8931],
          [0.7803, 0.5677, 0.8506, 0.9263]],

         [[0.3932, 0.3417, 0.7972, 0.5994],
          [0.4425, 0.3852, 0.2629, 0.3216],
          [0.0266, 0.6012, 0.3801, 0.7938],
          [0.2763, 0.8868, 0.9584, 0.3357]]],


        [[[0.6654, 0.2899, 0.1876, 0.5717],
          [0.4726, 0.1386, 0.7997, 0.1153],
          [0.2871, 0.5972, 0.2653, 0.7075],
          [0.2861, 0.8345, 0.8160, 0.8445]],

         [[0.9974, 0.9070, 0.6352, 0.8751],
          [0.9258, 0.5089, 0.2803, 0.2517],
          [0.6847, 0.4595, 0.5682, 0.1129],
          [0.4974, 0.7461, 0.3794, 0.1615]]]],requires_grad=True)
```

```python
x = m(input)
y = s(x)
z = y.sum()
print(x)
print(y)
print(z)
z.backward()
print(z.grad)
print(y.grad)
print(x.grad)
print(input.grad)
```
