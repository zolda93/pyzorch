```python
import zorch
from zorch import nn
```

```python
w = zorch.tensor([[[[-0.1415,  0.1721,  0.3044],
          [ 0.0890,  0.3689,  0.3387]]],


        [[[-0.2737, -0.1855,  0.1252],
          [-0.2890,  0.1660,  0.2861]]]],requires_grad=True)
bias = zorch.tensor([0.1634], requires_grad=True)
```

```python
m = nn.ConvTranspose2d(2,1,(2, 3), stride=(1, 2),padding=(1,1))
n = nn.MaxPool2d(2)
s = nn.Sigmoid()
m.weight = w
m.bias = bias
```

```python
input = zorch.Tensor([[[[0.5204, 0.6969, 0.3145, 0.7323],
          [0.3172, 0.1865, 0.7051, 0.3218],
          [0.9569, 0.1897, 0.9825, 0.9469],
          [0.5698, 0.9277, 0.9704, 0.6772]],

         [[0.1053, 0.5421, 0.0318, 0.1402],
          [0.6402, 0.4462, 0.2334, 0.9510],
          [0.0179, 0.0579, 0.8885, 0.2438],
          [0.8422, 0.8862, 0.9110, 0.8459]]],


        [[[0.4041, 0.2375, 0.1982, 0.4136],
          [0.3250, 0.6069, 0.7985, 0.4810],
          [0.6915, 0.2921, 0.7205, 0.6102],
          [0.7537, 0.5698, 0.5494, 0.7544]],

         [[0.1911, 0.9839, 0.6755, 0.5612],
          [0.5243, 0.8376, 0.8677, 0.4343],
          [0.0120, 0.6652, 0.6468, 0.8048],
          [0.8079, 0.7899, 0.7115, 0.4795]]]], requires_grad=True)
```

```python
x = m(input)
t = n(x)
y = s(t)
z = y.sum()
print(input)
print(x)
print(t)
print(y)
print(z)

z.backward()
print(z.grad)
print(y.grad)
print(t.grad)
print(x.grad)
print(m.weight.grad)
print(m.bias.grad)
print(input.grad)
```
