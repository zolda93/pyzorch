```python
import zorch
from zorch import nn
```

```python
w = zorch.tensor([[[[ 0.1505,  0.1984,  0.0946],
          [ 0.2708, -0.0505,  0.1905]],

         [[-0.1732,  0.0299,  0.1050],
          [-0.0840,  0.1170,  0.1909]]]],requires_grad=True)
bias = zorch.tensor([0.1072], requires_grad=True)
```

```python
m = nn.Conv2d(2,1,(2, 3), stride=(1, 2),padding=(1,1))
n = nn.MaxPool2d(2)
s = nn.Sigmoid()
m.weight = w
m.bias = bias
```

```python
input = zorch.Tensor([[[[0.8553, 0.8690, 0.0604, 0.6487],
          [0.8891, 0.2951, 0.2435, 0.6207],
          [0.4208, 0.1142, 0.9932, 0.2107],
          [0.3198, 0.4577, 0.1718, 0.9639]],

         [[0.7590, 0.1228, 0.7444, 0.9778],
          [0.3667, 0.3136, 0.5603, 0.4672],
          [0.9368, 0.2936, 0.4342, 0.0028],
          [0.2137, 0.5025, 0.1449, 0.5611]]],


        [[[0.9690, 0.9150, 0.0506, 0.6171],
          [0.2466, 0.5243, 0.1823, 0.2013],
          [0.0425, 0.4508, 0.1612, 0.5500],
          [0.0143, 0.4380, 0.7362, 0.9498]],

         [[0.2803, 0.8233, 0.5504, 0.5492],
          [0.3900, 0.1778, 0.6352, 0.9807],
          [0.4359, 0.0227, 0.7754, 0.3239],
          [0.6955, 0.4428, 0.1728, 0.6228]]]], requires_grad=True)
```

```python
x = m(input)
t = n(x)
y = s(t)
z = y.sum()
print(input)
print(x)
print(t)
print(y)
print(z)

z.backward()
print(z.grad)
print(y.grad)
print(t.grad)
print(x.grad)
print(m.weight.grad)
print(m.bias.grad)
print(input.grad)
```
